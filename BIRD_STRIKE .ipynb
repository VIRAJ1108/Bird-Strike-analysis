```python
df['FlightDate'] = pd.to_datetime(df['FlightDate'], format='%m/%d/%y %H:%M', errors='coerce')
```

```python
df['year'] = df['FlightDate'].dt.year
```

```python
strike=df.groupby('year')
```

```python
strike['Wildlife: Number Struck Actual'].sum() #2
```

```python
strike = df.groupby('year')['Wildlife: Number Struck Actual'].sum()
```

```python
plt.figure(figsize=(10, 6))
strike.plot(kind='bar',)
plt.title('Total Bird Strikes by YEARS ')
plt.xlabel('Years')
plt.ylabel('Total Number of Bird Strikes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
cost = df.groupby('year')
```

```python
cost['Cost: Total $'].sum() #5
```

```python
df['Wildlife: Number struck'] = pd.to_numeric(df['Wildlife: Number struck'], errors='coerce')
alti = df.groupby('Altitude bin')['Wildlife: Number struck'].sum()
```

```python
plt.figure(figsize=(10, 6))
alti.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Total Bird Strikes by Altitude ')
plt.xlabel('Altitude Bin')
plt.ylabel('Total Number of Bird Strikes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()#7
```

```python
phase1 = df.groupby('When: Phase of flight')
phase1['Wildlife: Number Struck Actual'].sum()
```

```python
df['Wildlife: Number struck'] = pd.to_numeric(df['Wildlife: Number struck'], errors='coerce')
phase = df.groupby('When: Phase of flight')['Wildlife: Number Struck Actual'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize=(10, 6))
phase.plot(kind='bar',)
plt.title('Total Bird Strikes by Phase of flight ')
plt.xlabel('Phases')
plt.ylabel('Total Number of Bird Strikes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()#8
```

```python
effect = df.groupby('Effect: Impact to flight')
```

```python
effect['Wildlife: Number Struck Actual'].sum()
```

```python
effect1 = df.groupby('Effect: Impact to flight')['Wildlife: Number Struck Actual'].sum().sort_values(ascending = False)
```

```python
plt.figure(figsize=(10, 6))
effect1.plot(kind='bar')
plt.title(' Impact on Flight')
plt.xlabel('Effect')
plt.ylabel('Number of bird strikes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()#10
```

```python
df['Altitude bin'] = df['Altitude bin'].astype(str)
df['Effect: Impact to flight'] = df['Effect: Impact to flight'].astype(str)

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Altitude bin'], df['Effect: Impact to flight'], alpha=0.5)
plt.title('Effect of Bird Strike by Altitude Bin')
plt.xlabel('Altitude Bin')
plt.ylabel('Effect of Strike')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
warning = df.groupby('Pilot warned of birds or wildlife?')['Wildlife: Number Struck Actual'].sum()
```

```python
plt.figure(figsize=(10, 6))
warning.plot(kind='bar')
plt.title(' Pilots Warning')
plt.xlabel('Effect')
plt.ylabel('Number of bird strikes')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
